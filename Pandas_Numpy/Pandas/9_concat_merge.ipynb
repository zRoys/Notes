```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'A':['A0', 'A1', 'A2','A3'],
        'B':['B0', 'B1', 'B2','B3'],
	'C':['C0', 'C1', 'C2','C3'],
	'D':['D0', 'D1', 'D2','D3']})


df2 = pd.DataFrame({'A':['A4', 'A5', 'A6','A7'],
	'B':['B4', 'B5', 'B6','B7'],
	'C':['C4', 'C5', 'C6','C7'],
	'D':['D4', 'D5', 'D6','D7']})
```

```python
# de esta manera cocncatenas dataframe, ingnoreindx haces que no se repitan los indices 
pd.concat([df1,df2],ignore_index=True)
```

```python
# de esta manera se concatena de manera de columnas muy raro 
pd.concat([df1,df2],axis=1)
```

## merge 

```python
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
 'A' : ['A0', 'A1', 'A2','A3'],
'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
 'C' : ['C0', 'C1', 'C2','C3'],
'D': ['D0', 'D1', 'D2','D3']})
```

```python
# de esta manera hacemos un join de un lado a otro
izq.merge(der, on='key')
```

```python
# solo se cambia de posicion 
der.merge(izq, on='key')
```

```python
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
 'A' : ['A0', 'A1', 'A2','A3'],
'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key_2' : ['k0', 'k1', 'k2','k3'],
 'C' : ['C0', 'C1', 'C2','C3'],
'D': ['D0', 'D1', 'D2','D3']})
```

```python
# de esta manera puedes hacer una union auntque no tengan el mismo nombre en las columnas 
izq.merge(der, left_on='key',right_on='key_2')
```

```python
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
 'A' : ['A0', 'A1', 'A2','A3'],
'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key_2' : ['k0', 'k1', 'k2',np.nan],
 'C' : ['C0', 'C1', 'C2','C3'],
'D': ['D0', 'D1', 'D2','D3']})
der
```

```python
# de esta manera puedes traer valores nulos sin que se pierdan en la concatenacion 
# con how='right' o how='left'  escoge la prioridad 
izq.merge(der, left_on='key',right_on='key_2',how='right')
```
