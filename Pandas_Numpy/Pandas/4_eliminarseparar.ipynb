```python
import pandas as pd
import numpy as np
```

```python
df_books = pd.read_csv('/home/roy/Recurses/bestsellers.csv',sep=',',header=0)
```

```python
df_books.head(2)
```

## Drop columns

```python
# de esta manera lo borras de la salida y no del dataframe
df_books.drop('Genre',axis=1).head(2)
```

```python
# de esta manera lo borras del dataframe
df_books.drop('Genre',axis=1,inplace=True)
```

```python
df_books.drop('Year',axis=1).head(2)
```

```python
# de esta manera tambien puedes borrar del data frame
del df_books['Price']
```

## Drop rows

```python
df_books.head(2)
```

```python
# de esta manera borras una fila pero no del dataframe 
df_books.drop(0,axis=0).head(5)
```

```python
# de esta manera borras varias lineas pero no del dataframe 
df_books.drop([1,2,3],axis=0).head(5)
```

```python
# puedes borrar un rango de filas 
df_books.drop(range(0,10),axis=0).head(5)
```

## Add Columns

```python
df_books.head(2)
```

```python
# de esta manera creas una nueva columna importando numpy 
df_books['Nueva columna'] = np.nan
df_books
```

```python
# hace un arreglo de valores consecutivos de 0 hasta el numero de filas del dataframe 
data = np.arange(0,df_books.shape[0])
```

```python
# agrego una nueva columna de llamada Rango que le da los valores de data
df_books['Rango'] = data
df_books.head(2)
```

## Add Rows

```python
# usar pd.concat para hacer filas

df = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
new_row = pd.DataFrame({'A': [5], 'B': [6]})

# Agregar la fila al final del DataFrame
# df = pd.concat([dataframe, datosque quiere añadir], ignore_index=True)
df = pd.concat([df, new_row], ignore_index=True)
print(df)
df.ndim
```
