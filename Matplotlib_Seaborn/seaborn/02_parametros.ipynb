```python
sns.displot(data=tips, x='total_bill')
plt.show
```

```python
sns.displot(data=tips, x='total_bill',y='tip')
plt.show
```

```python
sns.displot(data=tips, x='total_bill',y='tip',hue='sex')
plt.show
```

```python
# sintaxis basica = seaborn.tipo de grafica(data,x=,y=,que segmento hue)
sns.displot(data=tips, x='total_bill',hue='sex')
plt.show
```

```python
sns.displot(data=tips, x='total_bill',hue='sex',kind='kde',legend=True,palette='dark',alpha=.5)
plt.show
# data: especifica el conjunto de datos a utilizar.
# x: especifica la variable del eje x.
# hue: especifica la variable de segmentación.
# kind: especifica el tipo de trazado (en este caso, "kde" para el gráfico de densidad kernel).
# legend: indica si se desea mostrar la leyenda o no (True o False).
# palette: especifica la paleta de colores a utilizar.
# alpha: especifica la transparencia de los elementos trazados en el gráfico.
```
