```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
tips['weekend'] = tips.apply(lambda x : True if (x['day'] == 'Sun' or x['day'] == 'Sat')  else False,axis=1)
tips.head(10)
```

```python
# es un histograma 
sns.histplot(data=tips, x="tip", cumulative=False,hue='sex',stat='count',multiple='dodge')
plt.show()
# data: especifica el conjunto de datos a utilizar.
# x: especifica la variable del eje x.
# cumulative: indica si se desea un histograma acumulativo o no (True o False).
# hue: especifica la variable de segmentación.
# stat: indica la estadística a calcular para representar la altura de cada barra del histograma (en este caso, "count" para contar el número de observaciones en cada bin).
# multiple: especifica cómo se deben apilar las barras para múltiples variables de segmentación (en este caso, "dodge" para colocarlas lado a lado).
```

```python
# este es un diagrama de dencidad
sns.kdeplot(data = tips,x='tip',hue='sex',cumulative=False,fill=True,bw_adjust=1)
plt.show()
# data: especifica el conjunto de datos a utilizar.
# x: especifica la variable del eje x.
# hue: especifica la variable de segmentación.
# cumulative: indica si se desea una distribución de densidad acumulativa o no (True o False).
# fill: indica si se deben rellenar las áreas bajo la curva de densidad o no (True o False).
# bw_adjust: ajusta el ancho de banda de la estimación de densidad, que influye en la suavidad de la curva.
```

```python
# este es un grafico escalonado
sns.ecdfplot(data=tips,x='tip',hue='sex',stat='count')
plt.show()
```
