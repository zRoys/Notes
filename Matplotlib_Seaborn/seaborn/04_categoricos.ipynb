```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
# muestra la frecuencia de observaciones en cada categoría de una variable categórica en barras
sns.countplot(data=tips,x='day',hue='sex')
plt.show()
```

```python
plt.figure(figsize=(6,6))
# grafico tipo dispersion
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,color='0',marker='>')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,split= True,palette='pastel')
plt.show()
```

```python
sns.violinplot(data= tips, y = 'total_bill')
```

```python
# de esta manera puedes usar multiples graficos solo cambiendo en kind la variable 
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,kind='swarm',col='time')
plt.show()
# data=tips,    # Conjunto de datos que se va a trazar
# x='day',      # Variable para el eje x
# y='total_bill',  # Variable para el eje y
# hue='sex',    # Variable para segmentar los datos por color
# dodge=True,   # Indica si los puntos deben estar "esquivando" unos a otros
# kind='swarm', # Tipo de gráfico, en este caso "swarm" para un gráfico de abejas
# col='time'    # Variable categórica para crear múltiples subgráficos basados en las diferentes categorías
```
