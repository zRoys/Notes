```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
# de esta manera une dos maneras de visualizar los datos 
sns.jointplot(
    data=tips,          # especifica el conjunto de datos a utilizar
    x='total_bill',     # especifica la variable del eje x
    y='tip',            # especifica la variable del eje y
    hue='sex',          # especifica la variable de segmentación
    kind='hist',        # especifica el tipo de trama (hist, kde, etc.)
    marginal_ticks=True,# indica si se deben mostrar las marcas del eje marginal o no (True o False)
    marginal_kws=dict(
        bins=20,        # especifica el número de contenedores (bins) para el histograma marginal
        fill=False,     # indica si se deben rellenar las barras del histograma marginal o no (True o False)
        multiple='dodge'# especifica cómo se deben apilar las barras del histograma marginal (stack, dodge, etc.)
    )
)
```

```python
# pairplot funciona con las variable numericas y las une en varios graficos 
sns.pairplot(data=tips,hue='sex',palette='pastel',corner=True,kind='scatter')
plt.show()
```
