```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
# de esta manera puedes hacer un grafico de dispercion
sns.scatterplot(
    data=tips,     # Conjunto de datos que se va a trazar
    x='total_bill', # Variable para el eje x
    y='tip',       # Variable para el eje y
    hue='day',     # Variable para segmentar los datos por color
    style='time',  # Variable para segmentar los datos por forma
    size='size'    # Variable para segmentar los datos por tamao
)
# de esta manera posicionas la leyenda
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show
```

```python
plt.figure(figsize=(8,8))
markers = {"Lunch":"D","Dinner":"s"}
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size',markers=markers)
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show
```

```python
sns.lineplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size')
plt.show()
```

```python
# con replot puedes usar varias data visualitation de relacion 
plt.figure(figsize=(2,1))
sns.relplot(
    data=tips,         # Conjunto de datos que se va a trazar
    x='total_bill',    # Variable para el eje x
    y='tip',          # Variable para el eje y
    hue='day',        # Variable para segmentar los datos por color
    style='time',     # Variable para segmentar los datos por forma
    size='size',      # Variable para segmentar los datos por tamao
    kind='scatter',   # Tipo de grfico a trazar
    col='time'        # Variable para segmentar los datos en columnas
)
plt.show
```

```python
# de esta manera puedes hacer comparativas 
sns.pairplot(data=tips,hue='day')
plt.show() 
```
