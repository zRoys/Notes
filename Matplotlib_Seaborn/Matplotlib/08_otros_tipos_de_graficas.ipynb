```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
data = np.random.randint(1,50,100)
```

## Histograma

```python
plt.hist(data,bins=10,histtype='bar')
plt.show
```

```python
plt.hist(data,bins=10,histtype='step')
```

## Boxplit

```python
# plt.boxplot(data,orientacion,relleno,enfoca los datos en la mediana,veo solo la distribucion donde estan concentrados mis datos conFalse)
data = np.append(data,200)
plt.boxplot(data,vert=False,patch_artist=True,notch=True,showfliers=False)
plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30*np.random.rand(N)) **2
colors = np.random.rand(N)
```

```python
# de esta manera haces una tabla de dispercion con dos valores,s= para ver el area, c para color, marker para seleccionar marcadores, alpha para hacer transparente 
plt.scatter(x,y,s=area,c=colors,marker='o',alpha=0.5)
plt.show
```
