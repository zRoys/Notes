```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
# plt.subplot(filas,columnas,index)
plt.subplot(1,2,1)
plt.plot(x,y,'r:')
plt.plot(y,x,'b--')
plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'r:')
plt.plot(y,x,'b--')
plt.subplot(2,1,2)
plt.pie(y)
plt.show()
```

```python
# En este caso hicimos un grfico que contiene 2 filas y 3 columnas, 
# por lo que vamos a tener un grfico compuesto por 6 grficos.

# Con los ndices indicamos donde queremos que cada grfico se ejecute y 
# los ndices se ordenan de la siguiente forma:

# 1 2 3

# 4 5 6
```

```python
x= np.linspace(1,10,11,dtype=int)
y= x ** 2

plt.subplot(2,3,1)
plt.plot(x,y,'g:') # Grfico de lineas
plt.subplot(2,3,2)
plt.plot(x,y,'r--') # Grfico de lineas con puntos
plt.subplot(2,3,3)
plt.pie(x) # Grfico de torta
plt.subplot(2,3,4)
plt.boxplot(x) # Grfico de cajas
plt.subplot(2,3,5)
plt.bar(x,y) # Grfico de barras
plt.subplot(2,3,6)
plt.hist(x) # Grfico histograma
```
