```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,100,)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
# de este modo puedes crear 1 fila con 2 columnas 
# fig, axes = plt.subplots(nrows(numero de filas)=1,ncols(numero de columnas)=2)
fig, axes = plt.subplots(nrows=1,ncols=2)
# de esta manera puedes escoger con axes[numero] en que plano se traza el dibujo 
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
# es lo mismo solo que se definio en este caso al principio 
```

```python
fig, axes = plt.subplots(2,4) #no hace falta poner nrows=1,ncols=2
axes[0,0].plot(x,np.cos(x))
axes[0,1].plot(x,np.sin(x),'y')
axes[0,2].plot(x,np.tan(x),'r')
axes[0,3].plot(x,np.cos(x)**3,'g')
# de esta manera puedes asignar un grfico por posicion con axes[fila,columna]
```

```python
fig, ((axes1, axes2, axes3, axes4), (axes5, axes6, axes7, axes8)) = plt.subplots(2,4)
axes1.plot(x, np.sin(x), 'b')
axes2.plot(y, np.cos(x), 'purple')
axes3.plot(x, np.tan(y), 'orange')
axes4.plot(y, np.cos(y)**2, 'black')
axes5.plot(x*-1)
axes6.plot(y*-1)
axes7.plot(x**2,'gD')
axes8.plot(x,np.tan(x),'y')

#mejorar la visualizacin de las figuras
fig.tight_layout()
```
