```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(0,5,100)
y = np.sin(x)
```

```python
fig , axes = plt.subplots(1,2,figsize = (5,5)) #con figzise = cambias el tama√±o
# define los labels en la grafica y con legend se ponen, notacion matematica con $
axes[0].plot(x,y,label='$sin(x)$')
axes[0].set_title('Relacion x - y ') # de esta manera puedes poner titulos
axes[0].set_xlabel('x') # de esta manera puedes poner valores de los ejes x o y 
axes[0].set_ylabel('y')
axes[0].legend()

axes[1].plot(y,x)
axes[1].set_title('Relacion y - x ')
axes[1].set_xlabel('y')
axes[1].set_ylabel('x')
```

```python
plt.plot(x,y,label= 'sin(x)')
plt.title('Este es el comienzo')
plt.xlabel('x')
plt.ylabel('y')
# plt.legend(loc='lower left',bbox_to_anchor=(1,0.2)) de esta manera puedes 
# mover la posicion de la legenda 
plt.legend()
```
