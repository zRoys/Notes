```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
countrys = ["INDIA", "JAPAN", "MEXICO", "COLOMBIA", 'GERMANY']
population = [2000, 500, 900, 1000, 300]
```

```python
# plt.bar(countrys,population,width(ancho entre las barras)= numero,color = [por barras usa comas para definirlas])
plt.bar(countrys,population,width=0.5,color = ['red','blue'])
# de esta manera estableces los nombres de las etiquetas en el eje x, rotation para la rotacion de las etiquetas
plt.xticks(np.arange(5),('india','japon','mexico','colombia','alemania'),rotation = 45)
plt.show
```

```python
# de manera horizontal 
plt.barh(countrys,population)
plt.show
```

```python
x = ["Argentina", "Colombia", "Per", "Chile", "Bolivia", "Paraguay", "Ecuador", "Uruguay", "Venezuela"] # paises 
y = [47.33, 49.07, 32.16, 19.46, 11.05, 6.81, 18.00, 3.46, 28.52] # poblacion (data numrica)

plt.figure(figsize=(12, 6))
plt.bar(x, y, width=0.5, color=["tab:blue", "tab:orange", "tab:green", "tab:red", "tab:purple"])
plt.title("Poblacin en Millones de Habitantes") #titulo de la tabla
# por cada i,v en la enumeracion de 'y' que va a cer de 0-8 i va a ser cada numero de la enumeracion y v el valor de esa enumeracion 
# ejemplo i2 = v32.16
#  plt.text(i pon la enumeracion, 
#           v+.5 pon el valor .5 veces arriba , 
#           str(v)+' M' agrega M despues del valor v
#           , verticalalignment='bottom' alinealo de altura, 
#           horizontalalignment='center' alinealo al centro)
for i, v in enumerate(y):
    plt.text(i, v+.5, str(v)+' M',verticalalignment='bottom', horizontalalignment='center')

plt.xticks(np.arange(9),("ARGENTINA", "COLOMBIA", "PER", "CHILE", "BOLIVIA", "PARAGUAY", "ECUADOR", "URUGUAY", "VENEZUELA"), rotation=45)
plt.show()
```
