```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure() # aqui se creo un prano vacio 
axes = fig.add_axes([0.1,0.1,0.5,0.9]) # aqui a fig se le asignaron los axes 
#fig.add_axes([posicion x ,posicion y ,tamaño x ,tamaño y ])
axes.plot(x,y,'b') # aqui se puso que tipo de grafica y que valores lleva 
fig.show()
```

```python
# de esta manera pones dos o mas graficos en 1 mismo 
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes_2 = fig.add_axes([0.15,0.55,0.4,0.3])
#fig.add_axes([posicion x ,posicion y ,tamaño x ,tamaño y ])
axes.plot(x,y, 'b')
axes_2.plot(y,x,'r')
# aqui cambio el color del fondo 
axes_2.set_facecolor('#254649')
plt.show()
```

```python
# Un objeto define una figura, 
# esa figura es un lienzo en el cual podemos introducir diferentes graficas(axes)
# , de las cuales cada una posee sus propios ejes(axis).
```

```python
fig = plt.figure() #Objeto
axes = fig.add_axes([0.1,0.9, 0.5, 0.5])    #Grafica 1 del Objeto
axes2 = fig.add_axes([0.8, 0.9, 0.5, 0.5]) #Grafica 2 del Objeto
axes_3 = fig.add_axes([0.1,0.1,0.5,0.5])
axes_4 = fig.add_axes([0.8,0.1,0.5,0.5])
axes.plot(x,y,'b')   #Plot 1
axes2.plot(y,x,'r')  #Plot 2
axes_3.bar(x,y,color=['red'])
axes_4.pie(x)
axes2.set_facecolor('gray')
fig.show()
```
