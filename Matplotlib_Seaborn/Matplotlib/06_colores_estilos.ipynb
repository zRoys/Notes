## Matlab 

```python
# de esta manera puedes poner tu fondo preferido 
plt.style.use("dark_background")
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'b-')
ax.plot(x,x+3,'y.-')
ax.plot(x,x+4,'go:')
```

```python
# de esta manera puedes personalizar mejor tus visualgrafic con pyplot 
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,color= 'green',alpha=0.8,linewidth=5,linestyle='-',marker='o'
        ,markersize=10,markerfacecolor= "red")
ax.plot(x,x+2,color= 'blue',linewidth=8,linestyle='--',marker='1')
ax.plot(x,x+3,color= '#66FF66',linewidth=3,linestyle='dashed',marker='v')
ax.plot(x,x+4,color= '#CCFFCC',linewidth=2,linestyle=':',marker='P')
```

```python

fig, ax = plt.subplots(figsize=(6, 3.7))
ax.plot(x,x+1, label='circle', color = 'blue', alpha=0.8, linewidth=4,  linestyle='--', marker='8', markersize='10')
ax.plot(x,x+2, label='circle', color = 'blue', alpha=0.8, linewidth=4,  linestyle='--', marker='8', markersize='10')
ax.plot(x,x+3, label='hexagon', color = '#32CD32', alpha=0.8, linewidth=4, linestyle='dashed', marker='v', markersize='10')
ax.plot(x,x+4, label='cross',color = '#FF00FF', alpha=0.8, linewidth=4, linestyle=':', marker='P', markersize='10')
fig.legend(loc=(1,9),bbox_to_anchor=(1,0.2))
```
