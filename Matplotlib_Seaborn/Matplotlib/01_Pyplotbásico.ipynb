```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
# de esta manera puedes hacer una grafica con dos valores 
plt.plot(x,y)
plt.show
```

```python
# plt.plot(variable 1,variable 2,'letra del color tipo de visualisacion')
plt.plot(x,y,'bx')
plt.show
```

```python
plt.plot(x,y,'rs-')
plt.show
```

```python
plt.plot(x,y,'yD:')
plt.show
```

## despues de plt puedes poner el tipo de barra que quieras  

```python

plt.hist(x)
plt.show
```

```python
plt.pie(x)
plt.show
```

```python
plt.scatter(x,y)
plt.show
```

```python
plt.boxplot(x)
plt.show
```

```python
x = np.linspace(0, 8, 100)
y_sin = np.sin(x)
```

```python
plt.plot(x, y_sin,'r:', label='sin(x)')
plt.legend()
plt.show()
```
