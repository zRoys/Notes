```python
import pandas as pd 
import seaborn as sns 
import numpy as np 
import io 
```

la validación cruzada es una técnica que nos permite obtener una evaluación más precisa y generalizada del rendimiento de un modelo de aprendizaje automático al entrenarlo y evaluarlo varias veces con diferentes divisiones del conjunto de datos. Es especialmente útil cuando se tiene un conjunto de datos limitado y se quiere tener una idea más precisa de cómo se comportará el modelo en datos no vistos, la validación cruzada reduce el sesgo y la variabilidad asociados con una única partición entre datos de entrenamiento y prueba, pero no puede garantizar completamente la independencia de los resultados.

```python
df = pd.read_csv('/home/roy/Estadistica_inferencial/data/cancerdata_80367619-ee78-44cd-a7dc-7970bf56fa94.csv')
df
```

```python
X = df.iloc[:,2:] #variables numericas predictoras 
Y = df.iloc[:,1] # variable que queremos predecir 'categorica'
```

```python
X.head()
```

```python
Y.value_counts()
# como tenemos mas de 1 observacion podemos hacer este proceso del modelo de clasificacion 
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

# División de los datos en conjunto de entrenamiento y conjunto de prueba
# test_size=0.3 indica que el 30% de los datos se utilizará como conjunto de prueba,
# mientras que el 70% se utilizará como conjunto de entrenamiento.
# random_state=4 se utiliza para asegurar que se obtengan los mismos conjuntos de entrenamiento y prueba cada vez que se ejecute el código.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=4)

# Creación de una instancia del clasificador del árbol de decisión
modelo = DecisionTreeClassifier()

# Entrenamiento del modelo utilizando el conjunto de entrenamiento
modelo.fit(X_train, Y_train)

# Evaluación del rendimiento del modelo utilizando el conjunto de prueba
# El método score() calcula la precisión del modelo en los datos de prueba
resultado = modelo.score(X_test, Y_test)

print(resultado)
```

 el resultado obtenido muestra la precisión del modelo en el conjunto de prueba. En este caso, el modelo tuvo una precisión del 90.64%, lo que indica un buen rendimiento en términos de clasificación correcta de las muestras en el conjunto de prueba.

# validacion cruzada k fold

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

modelo = DecisionTreeClassifier()

# Creación de un objeto KFold para la validación cruzada con 10 folds (10 muestras)
KFold_validacion = KFold(10)

# Utilización de la validación cruzada para evaluar el rendimiento del modelo
resultados = cross_val_score(modelo, X, Y, cv=KFold_validacion)

# Imprimir los resultados de la validación cruzada para cada fold
print(resultados)
resultados.mean()
```

los resultados muestran la precisión del modelo en cada fold de la validación cruzada, y el promedio de estas puntuaciones proporciona una medida general del rendimiento del modelo. En este caso, el modelo tuvo una precisión promedio del 93.67%, lo que indica que es un modelo con un buen rendimiento en general.
