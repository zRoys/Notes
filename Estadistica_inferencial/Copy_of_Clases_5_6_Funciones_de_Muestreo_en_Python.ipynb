### 2) Muestreo sistemático:

 Sistemático: 
 
 Método de selección de ciertas unidades al azar y, a continuación, se eligen el resto siguiendo intervalos regulares. 
 
 Ejemplo: 

 El proceso de muestreo sistemático se puede explicar con un ejemplo sencillo. Imagina que tienes una lista de 1000 estudiantes y deseas seleccionar una muestra sistemática de 100 estudiantes. Primero, calculas el tamaño de paso, que es el número de elementos de la población dividido por el tamaño de la muestra (1000/100 = 10). Luego, seleccionas un número al azar entre 1 y 10, digamos el número 4.

Comenzando desde el número 4, seleccionas cada décimo estudiante en la lista original. En este caso, seleccionarías los estudiantes número 4, 14, 24, 34, y así sucesivamente hasta llegar al estudiante número 1000. De esta manera, obtienes una muestra sistemática que representa a la población de manera ordenada y equitativa.

El muestreo sistemático es útil cuando se tiene una lista o una secuencia ordenada de elementos y se desea obtener una muestra representativa. Al seleccionar elementos sistemáticamente, existe una mayor probabilidad de que la muestra refleje las características generales de la población. Además, el muestreo sistemático es más eficiente en términos de tiempo y recursos, ya que no requiere seleccionar elementos al azar repetidamente.

Sin embargo, es importante tener en cuenta que el muestreo sistemático puede presentar sesgos si existe algún patrón o regularidad en la lista original. Por ejemplo, si la lista de estudiantes está ordenada por alguna característica relevante, como el rendimiento académico, seleccionar sistemáticamente puede sesgar la muestra hacia un grupo particular de estudiantes.

```python


def systematic_sampling(econdata, step):
    # Generar un arreglo de índices utilizando la función np.arange().
    # El rango de índices va desde 0 hasta la longitud de econdata,
    # con un incremento de "step" en cada iteración en este caso va contando de 3 en 3.
    indexes = np.arange(0, len(econdata), step=step)
    
    # Utilizar los índices generados para obtener una muestra sistemática
    # de los datos económicos de econdata utilizando el método iloc[].
    systematic_sample = econdata.iloc[indexes]
    
    # Devolver la muestra sistemática generada.
    return systematic_sample

# Llamar a la función systematic_sampling con los parámetros econdata y step.
# Almacenar el resultado en la variable systematic_sample.
systematic_sample = systematic_sampling(econdata, 3)

systematic_sample
```

### 3) Muestreo estratificado:

• Estratificado: 

Método de selección de ciertas unidades por segmentos exclusivos y homogéneos y, a continuación, se elige una muestra aleatoria simple de cada segmento. Ejemplo: división por edades.

un ejemplo para ilustrar cómo funciona el muestreo estratificado: Supongamos que queremos estimar la calificación promedio de los estudiantes de una escuela y sabemos que hay tres grados: 9º, 10º y 11º. En lugar de tomar una muestra aleatoria simple de toda la población, podemos dividir la población en tres estratos basados en el grado y luego tomar una muestra de cada estrato proporcional al tamaño de cada uno.

Por ejemplo, supongamos que en la escuela hay 200 estudiantes de 9º grado, 300 estudiantes de 10º grado y 250 estudiantes de 11º grado. Si queremos una muestra de 100 estudiantes, podríamos tomar 30 estudiantes de 9º grado (30/200 * 100), 45 estudiantes de 10º grado (45/300 * 100) y 25 estudiantes de 11º grado (25/250 * 100). Esto nos daría una muestra estratificada donde cada grado está representado en proporción a su tamaño en la población.

El muestreo estratificado es útil cuando se desea obtener estimaciones más precisas para subgrupos específicos dentro de una población. Permite capturar las diferencias y heterogeneidad presentes en la población mediante la selección de muestras dentro de cada estrato.

Sin embargo, es importante asegurarse de que los estratos se definan de manera adecuada y que sean mutuamente excluyentes y exhaustivos. Además, se debe tener en cuenta el tamaño de cada estrato para asignar el tamaño de la muestra proporcionalmente.

1) Separa los datos por categorías (valores estratificados).

2) Define las categorías que quieres en tú muestro y su proporción.

3) Toma tantos datos como necesites de cada categoría para llegar a la proporción deseada.

4) Junta esas muestras de cada categoría para tu muestreo final.

Se crean dichas categorías a partir de la delegación y el tipo, esta información se almacena en la columna estratificado:

```python
econdata['estratificado'] = econdata['delegacion'] + "," + econdata['tipo']
(econdata['estratificado'].value_counts() / len(econdata)).sort_values(ascending=False)
```

Con esto ya tenemos los datos por categorías (paso 1) y podemos ver su proporción real:

Para nuestro muestreo definimos las siguientes proporciones (paso 2):

Cuautémoc, Hotel: 0.5,

Cuautémoc, Museo: 0.2,

Venustiano Carranza, Hotel: 0.1,

Cuauhtémoc, Mercado: 0.1,

Venustiano Carranza, Mercado: 0.1

Para crear dicha proporción a partir de los datos originales se utiliza la siguiente función que encuentra el número de muestras necesario de cada categoría (paso 3), hace un sampleo simple dentro de dicha categoria y luego junta esas muestras (paso 4):

```python
def data_estratificada(econdata, nombres_columnas_estrat, valores_estrat, prop_estrat, random_state=None):

    df_estrat = pd.DataFrame(columns = econdata.columns) # Creamos un data frame vacío con los nombres de las columnas de econdata

    pos = -1
    for i in range(len(valores_estrat)): # iteración sobre los valores estratificados
        pos += 1
        if pos == len(valores_estrat) - 1:
            ratio_len = len(econdata) - len(df_estrat) # si es la iteración final calcula el número de valores de salida tenga el mismo número de filas que de entrada
        else:
            ratio_len = int(len(econdata) * prop_estrat[i]) # calcula el número de filas según la proporción deseada

        df_filtrado = econdata[econdata[nombres_columnas_estrat] ==valores_estrat[i]] # filtra los datos de origen según los valores seleccionados en la estratificación de datos
        df_temp = df_filtrado.sample(replace=True, n=ratio_len, random_state=random_state) # haz un sample de los datos filtrados usando la ratio que hemos calculado

        df_estrat = pd.concat([df_estrat, df_temp]) # junta las tablas de sample con la estratificada para producir el resultado final

    return df_estrat # Return the stratified, re-sampled data
```

Utiliza este código para crear el sampleo:

```python
valores_estrat = ['Cuautémoc,Hotel', 'Cuautémoc,Museo', 'Venustiano Carranza,Hotel', 'Cuauhtémoc,Mercado','Venustiano Carranza,Mercado']
prop_estrat = [0.5, 0.2, 0.1, 0.1,0.1]
df_estrat = data_estratificada(econdata, 'estratificado', valores_estrat, prop_estrat, random_state=42)
df_estrat
```

Adicionalmente, puedes usar este código para ver la proporción final de tus datos:

En un análisis de datos, no hay un valor "ideal" específico para la varianza y la desviación estándar, ya que depende del contexto y los objetivos del análisis. Sin embargo, te puedo proporcionar algunas pautas generales que se suelen seguir.

En términos generales, una varianza baja y una desviación estándar baja indican que los datos están agrupados más cerca del promedio, lo cual puede ser beneficioso en algunos casos. Esto significa que los valores tienden a ser más consistentes y predecibles.

Por otro lado, una varianza alta y una desviación estándar alta indican que los datos están más dispersos alrededor del promedio. En este caso, los valores pueden ser más diversos y menos predecibles.

La elección de la varianza y la desviación estándar depende de varios factores, como la naturaleza de los datos, el dominio del problema y los objetivos del análisis. Por ejemplo, en algunas áreas, como las finanzas, se pueden preferir datos con menor variabilidad para reducir los riesgos. En otros casos, una mayor variabilidad puede ser aceptable o incluso deseable, como en estudios de investigación donde se exploran diferentes escenarios o resultados.

En resumen, no hay un valor "ideal" fijo para la varianza y la desviación estándar en un análisis de datos. La elección depende del contexto y los objetivos del análisis, y se deben considerar otros factores relevantes para tomar decisiones informadas.
