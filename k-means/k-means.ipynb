# ¿Qué es el algoritmo de K-means y cómo funciona?


Cantidad de clusters = k



busca reducir la distancia entre cada punto y su centroide al cuadrado con el fin de castigar la distancia si esta muy lejos 



se calcula la distancia de cada punto con respecto a cada centroide(estos son los centroides en un espacio c1=16 , c2=22) para ver su cercania y se le asigna un cluster a cada punto dependiendo de su cercania 



luego se vuelve a medir las distancias los valores obtenidos 



y asi hasta que no se muevan mas 

# graficamente 

se ponen centroides aleatorios en un plano 



cada uno de los puntos mide su distancia a los centroides 



y elige la distancia mas corta y se asigna a ese centroide, esto se hace con todos los puntos



se asignan los puntos a su centroide mas cercano y luego se recalculan los centroides utilizando los promedios que tenemos para el cluster



se reasignan los centroides y otravez se pregunta cada uno de los puntos su distancia respecto a los centroides cual es el mas cercano 



se reasignan y se recalcula una y otra vez hasta que no se muevan
