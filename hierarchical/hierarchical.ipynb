## Mtodo Ward

El Mtodo Ward se utiliza en el agrupamiento jerrquico de tipo Aglomerativo. Su criterio fundamental es el de mnima varianza, es decir, su objetivo es que tras agrupar y conseguir nuevos clusters, la varianza dentro de cada cluster sea la mnima posible. La funcin a minimizar es entonces la suma de las varianzas de cada cluster.

En cada iteracin, el algoritmo clusteriza de manera que se genere el menor aumento de varianza a la varianza total.

Este mtodo es utilizado como un criterio para indicar cmo se deben aglomerar los clusters en el proceso de agrupamiento jerrquico.
