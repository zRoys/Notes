## Método Ward

El Método Ward se utiliza en el agrupamiento jerárquico de tipo Aglomerativo. Su criterio fundamental es el de mínima varianza, es decir, su objetivo es que tras agrupar y conseguir nuevos clusters, la varianza dentro de cada cluster sea la mínima posible. La función a minimizar es entonces la suma de las varianzas de cada cluster.

En cada iteración, el algoritmo clusteriza de manera que se genere el menor aumento de varianza a la varianza total.

Este método es utilizado como un criterio para indicar cómo se deben aglomerar los clusters en el proceso de agrupamiento jerárquico.



## ¿Cuándo usar hierarchical clustering?


## Ventajas

- No necesitas conocer el número de clusters `K`.
- Es un método simple de implementar.
- El resultado es interpretable y fácil de entender.
- Solo requiere una ejecución para obtener el resultado.
- Proporciona una ayuda visual adicional mediante dendogramas.

## Desventajas

- Tiene un mal rendimiento en datasets largos debido a la complejidad computacional.
- No tiene un objetivo matemático claro para la formación de clusters, lo que puede llevar a resultados subjetivos.
- Los outliers en los datos pueden afectar negativamente el proceso de agrupamiento.
- Requiere un mayor poder de cómputo y recursos en comparación con otros algoritmos.

## ¿Cuándo usarlo?

Puedes considerar usar el agrupamiento jerárquico aglomerativo cuando:

- Deseas comprender los resultados visualmente mediante la representación de dendogramas.
- Tu dataset es pequeño o de tamaño moderado.
- Desconoces el número óptimo de clusters a priori.
- Necesitas obtener resultados rápidamente sin tener que preocuparte por ajustar parámetros adicionales.
