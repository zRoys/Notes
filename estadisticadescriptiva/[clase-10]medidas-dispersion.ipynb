## VARIANZA VS DESVIACION ESTANDAR 

 la varianza y la desviaci√≥n est√°ndar miden la dispersi√≥n o variabilidad de los datos, pero la varianza es el promedio de las diferencias al cuadrado entre cada dato y la media, mientras que la desviaci√≥n est√°ndar es simplemente la ra√≠z cuadrada de la varianza, lo que la hace m√°s f√°cil de interpretar y entender en t√©rminos de las unidades originales de los datos.

```python
# Desviaci√≥n estandar
# mide que tanto estan alejados los datos de la media y
# un numero mas cercano a 0 indica que estan menos alejados 
df['price_usd'].std()
```

```python
# Rango = valor max - valor min
# es la distancia de nuestros valores 
rango = df['price_usd'].max() - df['price_usd'].min()
rango
```

```python
# Quartiles
median = df['price_usd'].median()
Q1 = df['price_usd'].quantile(q=0.25)
Q3 = df['price_usd'].quantile(q=0.75)
min_val = df['price_usd'].quantile(q=0)
max_val = df['price_usd'].quantile(q=1.0)
print('min {} Q1 {} Q2 o median {} Q3 {} max {}'.format(min_val, Q1, median, Q3, max_val))
```

El IQR (rango intercuartil) es una medida de dispersi√≥n estad√≠stica que se utiliza para describir la distribuci√≥n de un conjunto de datos. **El IQR se define como la distancia entre el tercer cuartil (Q3) y el primer cuartil (Q1) de los datos. En otras palabras, el IQR es el rango de valores que se encuentra en el 50% intermedio de los datos.**

```python
iqr = Q3 - Q1
iqr
```

### L√≠mites para detecci√≥n de outliers (datos simetricamente distribuidos)

Datos entre $$Q_1 -1.5 \times \text{IQR}$$ y $$Q_3 + 1.5 \times \text{IQR}$$

Los outliers, tambi√©n conocidos como valores at√≠picos, son observaciones que se encuentran significativamente alejadas del resto de los datos en un conjunto de datos. Los outliers son valores que est√°n muy por encima o por debajo del rango t√≠pico de los datos y pueden influir en el an√°lisis estad√≠stico de los mismos.

```python
minlimit = Q1 - 1.5*iqr
maxlimit = Q3 + 1.5*iqr
print('rango para detecci√≥n de outliers: {}, {}'.format(minlimit, maxlimit))
```

```python
# Configurar el tama√±o predeterminado de las figuras de Seaborn
sns.set(rc={'figure.figsize':(11.7,8.27)})

# Crear una figura con dos subplots (subfiguras), uno para un histograma y otro para un boxplot
# `f` es la figura principal y `(ax_hist, ax_box)` son los dos subplots
# `2` indica que habr√° dos filas de subplots
# `sharex=True` indica que ambos subplots compartir√°n el mismo eje x
# `gridspec_kw={"height_ratios": (.6, .4)}` es un argumento que permite especificar la altura relativa de las dos subfiguras, en este caso `.6` para el histograma y `.4` para el boxplot
f, (ax_hist, ax_box) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.6, .4)})

# Crear un histograma utilizando Seaborn y lo asigna al subplot `ax_hist`
# `df['price_usd']` es la serie de datos que se va a graficar en el histograma
sns.histplot(df['price_usd'], ax=ax_hist)

# Crear un boxplot utilizando Seaborn y lo asigna al subplot `ax_box`
# `df['price_usd']` es la serie de datos que se va a graficar en el boxplot
sns.boxplot(df['price_usd'], ax=ax_box)

# Establecer una etiqueta vac√≠a para el eje x del subplot del histograma `ax_hist`
# Esto se hace para que no haya una etiqueta repetida en el eje x, ya que ambos subplots comparten el mismo eje x.
ax_hist.set(xlabel='')
```
