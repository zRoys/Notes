# Implementaci√≥n de la derivada

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x): # Funcion
  return x**2
x = np.linspace(-10,10, num=1000)

def df(f):# Funcion derivada
  h=0.00000000000000000000000000000000000001 #valor cercano a cero
  return (f(x+h)-f(x))/h # `f` es la funcion que se aplica
  #para entenderlo mejor si aplicas `np.cos` tambien funciona.
y = f(x)

plt.plot(x,y, 'b')
plt.plot(x,df(f), 'r')
plt.grid()
```

## Funci√≥n lineal
$$f(x)=mx+b$$

Derivada

$$f'(x)=m$$

```python
def f(x):
  return x

plt.plot(x, f(x), 'b')
plt.plot(x,df(f), 'r')
plt.grid()
```

## Funci√≥n escal√≥n o de Heaviside
$$H(x) =
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

Derivada

$$f¬¥(x)=0$$

```python
def H(x):
  Y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x>=0:
      Y[idx]=1
  return Y

N=1000


y = H(x)

plt.plot(x, H(x), 'b')
plt.plot(x,df(H), 'r')
plt.grid()
```

## Funci√≥n sigmoide

$$f(x)=\frac{1}{1-e^{-x}}$$

Derivada

$$f¬¥(x)=f(x)(1-f(x))$$

```python
def f(x):
  return 1/(1 + np.exp(-x))


N=1000

y = f(x)

plt.plot(x, f(x), 'b')
plt.plot(x,df(f), 'r')
plt.grid()
```

## Funci√≥n tangente hiperb√≥lica

$$f(x)=\frac{2}{1+e^{-2x}}-1$$

Derivada

$$f¬¥(x)=1-f(x)^2$$

```python
def f(x):
  return np.tanh(x)


N=1000

y = f(x)

plt.plot(x, f(x), 'b')
plt.plot(x,df(f), 'r')
plt.grid()
```

## Funci√≥n ReLU

$$R(x)=max(0,x)$$
 Tambi√©n se puede escribir como
$$R(x) =
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        x,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

Derivada

$$H(x) =
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
def f(x):
  return np.maximum(x,0)


N=1000

y = f(x)

plt.plot(x, f(x), 'b')
plt.plot(x,df(f), 'r')
plt.grid()
```
