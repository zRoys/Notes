## Escenarios a evitar en Machine Learning

Algo que debemos recordar es que nuestro modelo de Machine Learning puede caer en uno de los siguientes dos escenarios que debemos evitar:

### Underfitting

El Underfitting significa que nuestro modelo es demasiado simple y no logra captar adecuadamente los patrones presentes en los datos. En este caso, nuestro modelo no est adecuadamente ajustado a los features y la variable de salida. Para resolver el Underfitting, debemos investigar variables con mayor significado o considerar combinaciones y transformaciones de los datos para poder capturar de manera ms precisa la relacin entre los features y la variable de salida.

### Overfitting

El Overfitting, por otro lado, significa que nuestro modelo es demasiado complejo y se ajusta demasiado a los datos de entrenamiento. Sin embargo, esta alta complejidad hace que el modelo no generalice bien con datos del mundo real y tenga un rendimiento deficiente en situaciones nuevas. Para abordar el Overfitting, es necesario realizar una seleccin crtica de los features, descartando aquellos que no aportan informacin relevante o considerando combinaciones que retengan nicamente la informacin esencial para el problema en cuestin.

En ambos casos, es fundamental encontrar un equilibrio adecuado y evitar caer en extremos para obtener un modelo que sea capaz de generalizar de manera efectiva y realizar predicciones precisas en situaciones reales.


## Soluciones para evitar Underfitting y Overfitting

Para solucionar los problemas de Underfitting y Overfitting en nuestros modelos de Machine Learning, podemos aplicar las siguientes tcnicas:

### 1. Reduccin de la dimensionalidad con PCA

La reduccin de la dimensionalidad nos permite simplificar la cantidad de features manteniendo la mayor cantidad de informacin relevante. Utilizaremos el algoritmo de Anlisis de Componentes Principales (PCA) para transformar los features originales en un conjunto de componentes principales que capturen la mayor varianza de los datos. De esta manera, reduciremos la complejidad del modelo y evitaremos el Underfitting.

### 2. Regularizacin

La tcnica de la regularizacin consiste en penalizar aquellos features que no estn aportando informacin relevante o que podran estar afectando negativamente el rendimiento del modelo. La regularizacin puede ayudar a evitar el Overfitting al restringir los coeficientes de los features y evitar que se ajusten demasiado a los datos de entrenamiento.

### 3. Balanceo de datos

En problemas de clasificacin donde el conjunto de datos est desbalanceado, es decir, una clase tiene muchos ms ejemplos que otra, podemos aplicar tcnicas de balanceo como Oversampling y Undersampling. 

- **Oversampling:** Consiste en generar nuevas instancias de la clase minoritaria para igualar la cantidad de ejemplos de ambas clases.

- **Undersampling:** Consiste en eliminar instancias de la clase mayoritaria para igualar la cantidad de ejemplos de ambas clases.



Estas tcnicas ayudarn a mejorar el rendimiento de nuestro modelo al evitar que est sesgado hacia la clase mayoritaria.

Al aplicar estas soluciones, podremos obtener modelos ms equilibrados, que generalicen mejor y que sean ms efectivos en la prediccin de nuevos datos.


# PCA

# Introduccin al algoritmo PCA (Principal Component Analysis)

El algoritmo PCA, o Anlisis de Componentes Principales, es una tcnica ampliamente utilizada en Machine Learning para reducir la complejidad de un problema cuando se enfrenta a un dataset con una gran cantidad de features y relaciones complejas entre ellos y la variable que queremos predecir. Se utiliza para encontrar combinaciones lineales de las variables originales que contienen la mayor parte de la informacin relevante y permiten explicar la variabilidad de los datos.

## Por qu usaramos este algoritmo?

El algoritmo PCA se utiliza en situaciones donde:

1. Tenemos un nmero elevado de features y no todos son significativos para nuestro modelo.
2. Existe una alta correlacin entre los features, lo que puede introducir redundancia en el modelo.
3. El modelo sufre de Overfitting debido a la alta complejidad de los datos.
4. El costo computacional es alto y queremos reducir la cantidad de features para hacer el proceso ms eficiente.

## En qu consiste el algoritmo PCA

Bsicamente, el algoritmo PCA busca reducir la complejidad del problema en dos pasos:

1. Seleccionando solamente las variables relevantes.
2. Combinndolas en nuevas variables (componentes principales) que mantengan la informacin ms importante, es decir, la varianza de los features.



## Pasos para llevar a cabo el algoritmo PCA

Los pasos para aplicar el algoritmo PCA son los siguientes:

1. Calculamos la matriz de covarianza para expresar las relaciones entre nuestros features.
2. Hallamos los vectores propios y valores propios de esta matriz, que miden la fuerza y variabilidad de estas relaciones.
3. Ordenamos y seleccionamos los vectores propios con mayor variabilidad, ya que aportan ms informacin y nos permiten reducir la dimensionalidad del problema.

## Qu hacer si tenemos recursos limitados?

Si tenemos recursos limitados o un dataset demasiado exigente, podemos optar por variantes del PCA como:

1. **IPCA (Incremental PCA):** Utilizado para manejar grandes conjuntos de datos. Procesa los datos en bloques para reducir la carga computacional.
2. **KPCA (Kernel PCA):** Cuando nuestros datos no tienen una estructura separable linealmente, podemos aplicar un kernel que los mapee a un espacio de mayor dimensionalidad, donde el PCA pueda ser aplicado.

Estas variantes nos permiten adaptar el PCA a situaciones especficas y superar limitaciones computacionales.

El algoritmo PCA es una herramienta poderosa para la reduccin de dimensionalidad y mejora del rendimiento de nuestros modelos de Machine Learning en situaciones donde la cantidad de features es alta o los datos presentan complejas relaciones entre s.

# Alternativas al algoritmo PCA: Kernels

Una alternativa al algoritmo PCA es el uso de Kernels. Un Kernel es una funcin matemtica que toma mediciones que se comportan de manera no lineal y las proyecta en un espacio dimensional ms grande donde se vuelven linealmente separables.

## Para qu puede servir?

El uso de Kernels es til en casos donde los datos no son linealmente separables en el espacio original. Al aplicar un Kernel, los puntos son proyectados en otra dimensin donde se vuelven linealmente separables, lo que facilita el proceso de clasificacin o regresin.

Que tipo de funciones para Kernels nos podemos encontrar?



## Tipos de funciones de Kernels:

Existen diferentes tipos de funciones de Kernels que se utilizan en conjuntos de datos aplicados a clasificadores:

1. **Kernel Lineal:** Se utiliza para problemas linealmente separables y simplemente aplica una transformacin lineal a los datos.

2. **Kernel Polinomial:** Es til para mapear los datos en un espacio de mayor dimensionalidad utilizando funciones polinomiales.

3. **Kernel RBF (Radial Basis Function):** Es uno de los Kernels ms utilizados. Proyecta los datos en un espacio de infinitas dimensiones utilizando una funcin gaussiana.

4. **Kernel Sigmoidal:** Se utiliza para problemas de clasificacin binaria y emplea funciones sigmoidales para transformar los datos.

5. **Kernel Laplaciano:** Similar al RBF, pero utiliza una funcin Laplaciana para la proyeccin.

6. **Kernel Exponencial:** Otro tipo de Kernel basado en funciones exponenciales.



La eleccin del Kernel depender del tipo de problema que estemos abordando y de las caractersticas especficas de los datos. Cada Kernel tiene sus propias ventajas y desventajas, y puede tener un impacto significativo en el rendimiento del modelo de Machine Learning.

El uso de Kernels ofrece una alternativa poderosa y flexible para resolver problemas donde los datos no son linealmente separables y puede mejorar la capacidad de clasificacin o regresin de nuestros modelos.

# Regularizacin en Machine Learning

La tcnica de regularizacin consiste en disminuir la complejidad de nuestro modelo a travs de una penalizacin aplicada a las variables ms irrelevantes.

## Concepto de prdida

El concepto de prdida nos indica qu tan lejos estn nuestras predicciones de los datos reales. En general, buscamos minimizar la prdida para mejorar la calidad de nuestro modelo.



Como podemos apreciar en la grfica 1, hay un subajuste en el modelo representado por la lnea roja. La lnea se ajusta bien a los datos de entrenamiento, pero no generaliza bien para nuevos datos de prueba. Introducir un poco de sesgo a travs de la regularizacin puede ayudar a mejorar la varianza del modelo.



La grfica muestra cmo la prdida tiende a disminuir durante el entrenamiento del modelo. Sin embargo, llega un punto en el que la prdida empieza a aumentar para los nuevos datos de validacin, lo que indica sobreajuste. La prdida es la medida que utilizaremos para aplicar la regularizacin.

## Tipos de regularizacin

Existen dos tipos principales de regularizacin en Machine Learning:

### 1. L1 (Lasso):

- Reduce la complejidad eliminando features que no aportan demasiado al modelo.
- Penaliza los features poco relevantes volvindolos cero, eliminando el ruido que producen en el modelo.



### 2. L2 (Ridge):

- Reduce la complejidad disminuyendo el impacto de ciertos features en nuestro modelo.
- Penaliza los features poco relevantes, pero no los vuelve cero. Solamente limita la informacin que aportan a nuestro modelo.









### ElasticNet:

Hasta el momento hemos visto dos tcnicas de regularizacin que aaden un componente de penalizacin en el proceso de encontrar los valores de los parmetros  que minimizan la funcin de error.

## Regularizacin Lasso (L1) y Regularizacin Ridge (L2)

Recordando que :

* Ninguna de las dos es mejor que la otra para todos los casos.

* Lasso enva algunos coeficientes a cero permitiendo as seleccionar variables significativas para el modelo.

* Lasso funciona mejor si tenemos pocos predictores que influyen sobre el modelo.

* Ridge funciona mejor si es el caso contrario y tenemos una gran cantidad.

Para aplicarlos y decidir cul es el mejor en la prctica, podemos probar usando alguna tcnica como cross-validation iterativamente. o bien, podemos combinarlos

## Metodos de balanceo 

### Oversampling (Sobremuestreo): 

Consiste en aumentar la cantidad de ejemplos en las clases minoritarias mediante la duplicacin o generacin de datos sintticos. Esto ayuda a que las clases minoritarias tengan una mayor representacin en el conjunto de datos.

* Es til cuando tienes un conjunto de datos con una cantidad insuficiente de ejemplos en las clases minoritarias.

* Se recomienda cuando el tamao de la muestra es suficientemente grande y el sobremuestreo no conducir a un aumento significativo del tiempo de entrenamiento.

### Undersampling (Submuestreo): 

Implica reducir la cantidad de ejemplos en las clases mayoritarias para igualar el nmero de ejemplos en las clases minoritarias. Esto evita que las clases mayoritarias dominen el modelo.

* Es adecuado cuando tienes un conjunto de datos con una cantidad excesiva de ejemplos en las clases mayoritarias y deseas reducir el tamao de estas clases para equilibrar el conjunto de datos.

* Puede ser til cuando tienes restricciones computacionales y no puedes manejar un conjunto de datos demasiado grande.

### SMOTE (Synthetic Minority Over-sampling Technique): 

Es una tcnica de generacin sinttica de datos que crea ejemplos sintticos para las clases minoritarias mediante interpolacin entre ejemplos similares.

* Se recomienda cuando las clases minoritarias son escasas y necesitas aumentar su representacin en el conjunto de datos.

* Es especialmente til cuando no quieres perder informacin al reducir la clase mayoritaria mediante el undersampling.

### ADASYN (Adaptive Synthetic Sampling): 

Similar a SMOTE, pero da ms peso a los ejemplos de las clases minoritarias que son ms difciles de clasificar.

* Es una variante de SMOTE que puede ser til cuando las clases minoritarias estn altamente desbalanceadas y algunas de ellas son ms difciles de clasificar que otras.

* ADASYN se enfoca en generar ms ejemplos para las clases minoritarias que son ms difciles de clasificar, lo que puede mejorar el rendimiento del modelo en escenarios desafiantes.

### Ensemble de clasificadores: 

Puedes utilizar tcnicas de ensemble como el Bagging o Boosting con clasificadores que ya se ocupan del desequilibrio de clases, como Random Forest o Gradient Boosting Machines.

* Las tcnicas de ensemble como Bagging o Boosting pueden ser beneficiosas cuando se combinan con clasificadores que ya estn diseados para manejar el desequilibrio de clases, como Random Forest o Gradient Boosting Machines.

* Estos enfoques pueden mejorar el rendimiento del modelo al combinar las predicciones de varios clasificadores, lo que puede ser til en problemas con clases desbalanceadas.

### Modificar la funcin de prdida: 

Algunos algoritmos de aprendizaje profundo permiten modificar la funcin de prdida para penalizar errores en las clases minoritarias, lo que puede ayudar a abordar el desequilibrio.

* Esta opcin es relevante para algoritmos de aprendizaje profundo donde se puede ajustar la funcin de prdida para penalizar los errores en las clases minoritarias.

* Puede ser til cuando deseas adaptar el entrenamiento del modelo para centrarse ms en la clasificacin correcta de las clases minoritarias.
