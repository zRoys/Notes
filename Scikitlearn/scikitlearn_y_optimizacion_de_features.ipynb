## ¿Cómo afectan nuestros features a los modelos de Machine Learning?

Hacer una buena selección de nuestros features hará que nuestros algoritmos corran de una manera más eficiente. Una de las formas de saber que nuestros features han sido bien seleccionados es con el sesgo y la varianza.

**Sesgo y Varianza:**

Una mala selección de nuestros features nos puede llevar a alguno de estos dos escenarios indeseados.


## Escenarios a evitar en Machine Learning

Algo que debemos recordar es que nuestro modelo de Machine Learning puede caer en uno de los siguientes dos escenarios que debemos evitar:

### Underfitting

El Underfitting significa que nuestro modelo es demasiado simple y no logra captar adecuadamente los patrones presentes en los datos. En este caso, nuestro modelo no está adecuadamente ajustado a los features y la variable de salida. Para resolver el Underfitting, debemos investigar variables con mayor significado o considerar combinaciones y transformaciones de los datos para poder capturar de manera más precisa la relación entre los features y la variable de salida.

### Overfitting

El Overfitting, por otro lado, significa que nuestro modelo es demasiado complejo y se ajusta demasiado a los datos de entrenamiento. Sin embargo, esta alta complejidad hace que el modelo no generalice bien con datos del mundo real y tenga un rendimiento deficiente en situaciones nuevas. Para abordar el Overfitting, es necesario realizar una selección crítica de los features, descartando aquellos que no aportan información relevante o considerando combinaciones que retengan únicamente la información esencial para el problema en cuestión.

En ambos casos, es fundamental encontrar un equilibrio adecuado y evitar caer en extremos para obtener un modelo que sea capaz de generalizar de manera efectiva y realizar predicciones precisas en situaciones reales.


## Soluciones para evitar Underfitting y Overfitting

Para solucionar los problemas de Underfitting y Overfitting en nuestros modelos de Machine Learning, podemos aplicar las siguientes técnicas:

### 1. Reducción de la dimensionalidad con PCA

La reducción de la dimensionalidad nos permite simplificar la cantidad de features manteniendo la mayor cantidad de información relevante. Utilizaremos el algoritmo de Análisis de Componentes Principales (PCA) para transformar los features originales en un conjunto de componentes principales que capturen la mayor varianza de los datos. De esta manera, reduciremos la complejidad del modelo y evitaremos el Underfitting.

### 2. Regularización

La técnica de la regularización consiste en penalizar aquellos features que no están aportando información relevante o que podrían estar afectando negativamente el rendimiento del modelo. La regularización puede ayudar a evitar el Overfitting al restringir los coeficientes de los features y evitar que se ajusten demasiado a los datos de entrenamiento.

### 3. Balanceo de datos

En problemas de clasificación donde el conjunto de datos está desbalanceado, es decir, una clase tiene muchos más ejemplos que otra, podemos aplicar técnicas de balanceo como Oversampling y Undersampling. 

- **Oversampling:** Consiste en generar nuevas instancias de la clase minoritaria para igualar la cantidad de ejemplos de ambas clases.

- **Undersampling:** Consiste en eliminar instancias de la clase mayoritaria para igualar la cantidad de ejemplos de ambas clases.



Estas técnicas ayudarán a mejorar el rendimiento de nuestro modelo al evitar que esté sesgado hacia la clase mayoritaria.

Al aplicar estas soluciones, podremos obtener modelos más equilibrados, que generalicen mejor y que sean más efectivos en la predicción de nuevos datos.


# PCA

# Introducción al algoritmo PCA (Principal Component Analysis)

El algoritmo PCA, o Análisis de Componentes Principales, es una técnica ampliamente utilizada en Machine Learning para reducir la complejidad de un problema cuando se enfrenta a un dataset con una gran cantidad de features y relaciones complejas entre ellos y la variable que queremos predecir. Se utiliza para encontrar combinaciones lineales de las variables originales que contienen la mayor parte de la información relevante y permiten explicar la variabilidad de los datos.

## ¿Por qué usaríamos este algoritmo?

El algoritmo PCA se utiliza en situaciones donde:

1. Tenemos un número elevado de features y no todos son significativos para nuestro modelo.
2. Existe una alta correlación entre los features, lo que puede introducir redundancia en el modelo.
3. El modelo sufre de Overfitting debido a la alta complejidad de los datos.
4. El costo computacional es alto y queremos reducir la cantidad de features para hacer el proceso más eficiente.

## En qué consiste el algoritmo PCA

Básicamente, el algoritmo PCA busca reducir la complejidad del problema en dos pasos:

1. Seleccionando solamente las variables relevantes.
2. Combinándolas en nuevas variables (componentes principales) que mantengan la información más importante, es decir, la varianza de los features.



## Pasos para llevar a cabo el algoritmo PCA

Los pasos para aplicar el algoritmo PCA son los siguientes:

1. Calculamos la matriz de covarianza para expresar las relaciones entre nuestros features.
2. Hallamos los vectores propios y valores propios de esta matriz, que miden la fuerza y variabilidad de estas relaciones.
3. Ordenamos y seleccionamos los vectores propios con mayor variabilidad, ya que aportan más información y nos permiten reducir la dimensionalidad del problema.

## ¿Qué hacer si tenemos recursos limitados?

Si tenemos recursos limitados o un dataset demasiado exigente, podemos optar por variantes del PCA como:

1. **IPCA (Incremental PCA):** Utilizado para manejar grandes conjuntos de datos. Procesa los datos en bloques para reducir la carga computacional.
2. **KPCA (Kernel PCA):** Cuando nuestros datos no tienen una estructura separable linealmente, podemos aplicar un kernel que los mapee a un espacio de mayor dimensionalidad, donde el PCA pueda ser aplicado.

Estas variantes nos permiten adaptar el PCA a situaciones específicas y superar limitaciones computacionales.

El algoritmo PCA es una herramienta poderosa para la reducción de dimensionalidad y mejora del rendimiento de nuestros modelos de Machine Learning en situaciones donde la cantidad de features es alta o los datos presentan complejas relaciones entre sí.

## Kernels y KPCA

# Alternativas al algoritmo PCA: Kernels

Una alternativa al algoritmo PCA es el uso de Kernels. Un Kernel es una función matemática que toma mediciones que se comportan de manera no lineal y las proyecta en un espacio dimensional más grande donde se vuelven linealmente separables.

## ¿Para qué puede servir?

El uso de Kernels es útil en casos donde los datos no son linealmente separables en el espacio original. Al aplicar un Kernel, los puntos son proyectados en otra dimensión donde se vuelven linealmente separables, lo que facilita el proceso de clasificación o regresión.

### Ejemplo:

En la siguiente imagen, podemos observar dos conjuntos de datos que no son linealmente separables en el espacio bidimensional:



Sin embargo, si aplicamos un Kernel y proyectamos los datos en un espacio tridimensional, los puntos se vuelven linealmente separables

¿Que tipo de funciones para Kernels nos podemos encontrar?



## Tipos de funciones de Kernels:

Existen diferentes tipos de funciones de Kernels que se utilizan en conjuntos de datos aplicados a clasificadores:

1. **Kernel Lineal:** Se utiliza para problemas linealmente separables y simplemente aplica una transformación lineal a los datos.

2. **Kernel Polinomial:** Es útil para mapear los datos en un espacio de mayor dimensionalidad utilizando funciones polinomiales.

3. **Kernel RBF (Radial Basis Function):** Es uno de los Kernels más utilizados. Proyecta los datos en un espacio de infinitas dimensiones utilizando una función gaussiana.

4. **Kernel Sigmoidal:** Se utiliza para problemas de clasificación binaria y emplea funciones sigmoidales para transformar los datos.

5. **Kernel Laplaciano:** Similar al RBF, pero utiliza una función Laplaciana para la proyección.

6. **Kernel Exponencial:** Otro tipo de Kernel basado en funciones exponenciales.



La elección del Kernel dependerá del tipo de problema que estemos abordando y de las características específicas de los datos. Cada Kernel tiene sus propias ventajas y desventajas, y puede tener un impacto significativo en el rendimiento del modelo de Machine Learning.

El uso de Kernels ofrece una alternativa poderosa y flexible para resolver problemas donde los datos no son linealmente separables y puede mejorar la capacidad de clasificación o regresión de nuestros modelos.

## ¿Qué es la regularización y cómo aplicarla?

# Regularización en Machine Learning

La técnica de regularización consiste en disminuir la complejidad de nuestro modelo a través de una penalización aplicada a las variables más irrelevantes.

## Concepto de pérdida

El concepto de pérdida nos indica qué tan lejos están nuestras predicciones de los datos reales. En general, buscamos minimizar la pérdida para mejorar la calidad de nuestro modelo.



Como podemos apreciar en la gráfica 1, hay un subajuste en el modelo representado por la línea roja. La línea se ajusta bien a los datos de entrenamiento, pero no generaliza bien para nuevos datos de prueba. Introducir un poco de sesgo a través de la regularización puede ayudar a mejorar la varianza del modelo.



La gráfica muestra cómo la pérdida tiende a disminuir durante el entrenamiento del modelo. Sin embargo, llega un punto en el que la pérdida empieza a aumentar para los nuevos datos de validación, lo que indica sobreajuste. La pérdida es la medida que utilizaremos para aplicar la regularización.

## Tipos de regularización

Existen dos tipos principales de regularización en Machine Learning:

### 1. L1 (Lasso):

- Reduce la complejidad eliminando features que no aportan demasiado al modelo.
- Penaliza los features poco relevantes volviéndolos cero, eliminando el ruido que producen en el modelo.



### 2. L2 (Ridge):

- Reduce la complejidad disminuyendo el impacto de ciertos features en nuestro modelo.
- Penaliza los features poco relevantes, pero no los vuelve cero. Solamente limita la información que aportan a nuestro modelo.









### ElasticNet:

Hasta el momento hemos visto dos técnicas de regularización que añaden un componente de penalización en el proceso de encontrar los valores de los parámetros β que minimizan la función de error.

## Regularización Lasso (L1) y Regularización Ridge (L2)

Recordando que :

* Ninguna de las dos es mejor que la otra para todos los casos.

* Lasso envía algunos coeficientes a cero permitiendo así seleccionar variables significativas para el modelo.

* Lasso funciona mejor si tenemos pocos predictores que influyen sobre el modelo.

* Ridge funciona mejor si es el caso contrario y tenemos una gran cantidad.

Para aplicarlos y decidir cuál es el mejor en la práctica, podemos probar usando alguna técnica como cross-validation iterativamente. o bien, podemos combinarlos…

## Regularización ElasticNet

Es común encontrarnos en la literatura con un enfoque intermedio llamado ElasticNet, que combina las penalizaciones L1 y L2 en una sola función de optimización:

![pasted image 0-3.png](<attachment:pasted image 0-3.png>)

En esta ecuación, tenemos un parámetro adicional α (alfa) que varía entre 0 y 1. Si α = 0, ElasticNet se comporta como Ridge, y si α = 1, se comporta como Lasso. Así, ElasticNet nos brinda un espectro lineal de posibles combinaciones entre estas dos técnicas, lo que nos permite probar tanto L1 como L2 al mismo tiempo sin perder información.

ElasticNet supera las limitaciones individuales de Lasso y Ridge y puede ser la opción preferente para probar la regularización cuando se requiere una mayor flexibilidad o cuando no se tiene un conocimiento matemático profundo.

En la práctica, podemos utilizar técnicas como la validación cruzada para decidir cuál de estas técnicas de regularización (Lasso, Ridge o ElasticNet) se adapta mejor a nuestros datos y problemas específicos.


## Lasso vs Ridge

1. No hay un campeón definitivo para todos los problemas.
2. Si hay pocos features que se relacionen directamente con la variable a predecir, se puede probar Lasso.
3. Si hay varios features relacionados con la variable a predecir, se puede probar Ridge.

La elección entre Lasso y Ridge dependerá del contexto específico y la naturaleza de los datos.

La regularización es una herramienta valiosa para mejorar la generalización y el rendimiento de nuestros modelos de Machine Learning, evitando el sobreajuste y seleccionando los features más relevantes para nuestras predicciones.

## Metodos de balanceo 

### Oversampling (Sobremuestreo): 

Consiste en aumentar la cantidad de ejemplos en las clases minoritarias mediante la duplicación o generación de datos sintéticos. Esto ayuda a que las clases minoritarias tengan una mayor representación en el conjunto de datos.

* Es útil cuando tienes un conjunto de datos con una cantidad insuficiente de ejemplos en las clases minoritarias.

* Se recomienda cuando el tamaño de la muestra es suficientemente grande y el sobremuestreo no conducirá a un aumento significativo del tiempo de entrenamiento.

### Undersampling (Submuestreo): 

Implica reducir la cantidad de ejemplos en las clases mayoritarias para igualar el número de ejemplos en las clases minoritarias. Esto evita que las clases mayoritarias dominen el modelo.

* Es adecuado cuando tienes un conjunto de datos con una cantidad excesiva de ejemplos en las clases mayoritarias y deseas reducir el tamaño de estas clases para equilibrar el conjunto de datos.

* Puede ser útil cuando tienes restricciones computacionales y no puedes manejar un conjunto de datos demasiado grande.

### SMOTE (Synthetic Minority Over-sampling Technique): 

Es una técnica de generación sintética de datos que crea ejemplos sintéticos para las clases minoritarias mediante interpolación entre ejemplos similares.

* Se recomienda cuando las clases minoritarias son escasas y necesitas aumentar su representación en el conjunto de datos.

* Es especialmente útil cuando no quieres perder información al reducir la clase mayoritaria mediante el undersampling.

### ADASYN (Adaptive Synthetic Sampling): 

Similar a SMOTE, pero da más peso a los ejemplos de las clases minoritarias que son más difíciles de clasificar.

* Es una variante de SMOTE que puede ser útil cuando las clases minoritarias están altamente desbalanceadas y algunas de ellas son más difíciles de clasificar que otras.

* ADASYN se enfoca en generar más ejemplos para las clases minoritarias que son más difíciles de clasificar, lo que puede mejorar el rendimiento del modelo en escenarios desafiantes.

### Ensemble de clasificadores: 

Puedes utilizar técnicas de ensemble como el Bagging o Boosting con clasificadores que ya se ocupan del desequilibrio de clases, como Random Forest o Gradient Boosting Machines.

* Las técnicas de ensemble como Bagging o Boosting pueden ser beneficiosas cuando se combinan con clasificadores que ya están diseñados para manejar el desequilibrio de clases, como Random Forest o Gradient Boosting Machines.

* Estos enfoques pueden mejorar el rendimiento del modelo al combinar las predicciones de varios clasificadores, lo que puede ser útil en problemas con clases desbalanceadas.

### Modificar la función de pérdida: 

Algunos algoritmos de aprendizaje profundo permiten modificar la función de pérdida para penalizar errores en las clases minoritarias, lo que puede ayudar a abordar el desequilibrio.

* Esta opción es relevante para algoritmos de aprendizaje profundo donde se puede ajustar la función de pérdida para penalizar los errores en las clases minoritarias.

* Puede ser útil cuando deseas adaptar el entrenamiento del modelo para centrarse más en la clasificación correcta de las clases minoritarias.
