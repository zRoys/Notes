# Implementación de Batch K-Means

se sabe cuantos clusters se quiere tener 

```python
import pandas as pd 
import numpy as np 

from sklearn.cluster import MiniBatchKMeans

df = pd.read_csv('/home/roy/Scikitlearn/data/candy.csv')
df
```

```python
X = df.drop(columns=('competitorname'),axis=1)

kmeans = MiniBatchKMeans(n_clusters=4,batch_size=8).fit(X)
print('centros:', len(kmeans.cluster_centers_))
```

```python
print(kmeans.predict(X))
df['group'] = kmeans.predict(X)
```

```python
df
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
sns.pairplot(df[['sugarpercent','pricepercent','winpercent','group']], hue = 'group',palette='colorblind')
```

# Implementación de Mean-Shift

no se sabe el numero de clusters 


```python
from sklearn.cluster import MeanShift
```

```python
meanshift = MeanShift().fit(X)#bandwidth podriamos especificarlo pero tiene una tecnica matematica ya especificada para indentificar cual es el ancho de banda mas ajustado para nuestros datos 

print(max(meanshift.labels_)) # usar max para saber cual es el numero maximo de todo el areglo 
```

```python
print(meanshift.cluster_centers_)
```

```python
df['meanshift'] = meanshift.labels_
df
```

```python
sns.pairplot(df[['sugarpercent','pricepercent','winpercent','meanshift']], hue = 'meanshift',palette='colorblind')
```
