```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import norm 
```

## Distribución normal teórica


$$P(X) = \frac{1}{\sigma \sqrt{2 \pi}} \exp{\left[-\frac{1}{2}\left(\frac{X-\mu}{\sigma} \right)^2 \right]}$$

* $\mu$: media de la distribución
* $\sigma$: desviación estándar de la distribución

```python
# definimos nuestra distribución gaussiana, mu = media, sigma = desviasion estandar 
def gaussian(x, mu, sigma):
  return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-0.5*pow((x-mu)/sigma,2))
```

```python
x = np.arange(-4,4,0.1)
y = gaussian(x, 1.0, 0.5) # la desviasion estandar es la longitud del area bajo la curva 

plt.plot(x, y)
```

```python
# usando scipy
dist = norm(0, 1) # norm es hacer una distribucion gauseana, norm(mu,std)
x = np.arange(-4,4,0.1)
y = [dist.pdf(value) for value in x] # probabilidad distributiva 
plt.plot(x, y)
```

```python
# calculando la distribución acumulada
dist = norm(0, 1)
x = np.arange(-4,4,0.1)
y = [dist.cdf(value) for value in x] #probabilidad acumulada 
plt.plot(x, y)
```

## Distribución normal (gausiana) a partir de los datos

* *El archivo excel* lo puedes descargar en esta página: https://seattlecentral.edu/qelp/sets/057/057.html

```python
df = pd.read_excel('s057.xls')
arr = df['Normally Distributed Housefly Wing Lengths'].values[4:] #values es para que solo me de los valores 
values, dist = np.unique(arr, return_counts=True) # calcula de arr cuantas veces se repite un numero 
print(values) #valores
print(dist) # numero de veces que se repite cada valor 
plt.bar(values,dist)
```

```python
# estimación de la distribución de probabilidad media del arr
#distribución teórica
mu = arr.mean()
sigma = arr.std()
dist = norm(mu, sigma)
x = np.arange(30,60,0.1)
y = [dist.pdf(value) for value in x]
plt.plot(x, y,'r')

# datos
values, dist = np.unique(arr, return_counts=True)
plt.bar(values, dist/len(arr))
```
